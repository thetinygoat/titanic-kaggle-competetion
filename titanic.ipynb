```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv("train.csv")
data.head()

```

```python
data.columns
```

```python
data.drop(['PassengerId','Name', 'Ticket'], axis=1)
```

```python
sex = pd.get_dummies(data['Sex'], drop_first=True)
```

```python
emb = pd.get_dummies(data['Embarked'])
```

```python
new_data = data.drop(['PassengerId','Name', 'Ticket', 'Sex', 'Embarked', 'Cabin', 'Age'], axis=1)
```

```python
cleaned = pd.concat([new_data,sex,emb], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x = cleaned.drop('Survived', axis=1)
y = cleaned['Survived']
```

```python
rfc = RandomForestClassifier()
rfc.fit(x, y)
```

```python
test = pd.read_csv('test.csv')
```

```python
test.drop(['PassengerId','Name', 'Ticket', 'Sex', 'Embarked', 'Cabin', 'Age'], axis=1)
```

```python
sex_test = pd.get_dummies(test['Sex'], drop_first=True)
emb_test = pd.get_dummies(test['Embarked'])
new_test = test.drop(['PassengerId','Name', 'Ticket', 'Sex', 'Embarked', 'Cabin', 'Age'], axis=1)
cleaned_test = pd.concat([new_test,sex_test,emb_test], axis=1)
```

```python
pred = rfc.predict(cleaned_test.fillna(cleaned_test.mean()))
```
